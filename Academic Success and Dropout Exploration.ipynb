### Feature Engineering

```python
# Approval (passing) rates
first_sem_approval_rate = X['Curricular units 1st sem (approved)'].sum() / X['Curricular units 1st sem (enrolled)'].sum()
second_sem_approval_rate = X['Curricular units 2nd sem (approved)'].sum() / X['Curricular units 2nd sem (enrolled)'].sum()

print(f'The approval (passing) rate in the first semester is {first_sem_approval_rate}')
print(f'The approval (passing) rate in the second semester is {second_sem_approval_rate}')
```

```python
# Overall approval rate
overall_approval_rate = (X['Curricular units 1st sem (approved)'].sum() + X['Curricular units 2nd sem (approved)'].sum()) / (X['Curricular units 1st sem (enrolled)'].sum() + X['Curricular units 2nd sem (enrolled)'].sum())

print(f'The overall approval (passing) rate is {overall_approval_rate}')
```

```python
# Grade trend
grade_change = X['Curricular units 2nd sem (grade)'] - X['Curricular units 1st sem (grade)']

print(f'The average grade trend from the first to second semester is {grade_change.mean()}')
```

```python
# Financial risk
financial_risk_amount = ((X['Debtor'] == 1) & (X['Tuition fees up to date'] == 0)).astype(int).sum()
print(f'Number of people in financial risk: {financial_risk_amount}')

financial_risk_rate = financial_risk_amount / X.shape[0]
print(f'The financial risk rate is {financial_risk_rate}')
```

```python
# Scholarships
scholarship_amount = X['Scholarship holder'].sum()
scholarship_rate = scholarship_amount / X.shape[0]

print(f'The scholarship rate is {scholarship_rate}')
```

```python
# Scholarship with no debt
scholarship_no_debt_amount = ((X['Scholarship holder'] == 1) & (X['Debtor'] == 0)).astype(int).sum()
print(f'Number of students with scholarship and no debt: {scholarship_no_debt_amount}')

scholarship_no_debt_rate = scholarship_no_debt_amount / X.shape[0]
print(f'The scholarship rate for students without debt is {scholarship_no_debt_rate}')
```

```python
# Scholarship with debt
scholarship_and_debt_amount = ((X['Scholarship holder'] == 1) & (X['Debtor'] == 1)).astype(int).sum()
print(f'Number of students with scholarship and debt: {scholarship_and_debt_amount}')

scholarship_and_debt_rate = scholarship_and_debt_amount / X.shape[0]
print(f'The scholarship rate for students with debt is {scholarship_and_debt_rate}')
```

```python
# Age (older students)
older_student_amount = (X['Age at enrollment'] > 25).astype(int).sum()
print(f'Number of students older than 25: {older_student_amount}')

older_student_rate = older_student_amount / X.shape[0]
print(f'The older student rate is {older_student_rate}')
```
